```python
m = []
n = []
for i in all_stock_names:
    element1 = np.sum(df[i] < 0)
    m.append(element1)
    element2 = np.sum(df[i] > 0)
    n.append(element2)
             
```

```python
names = []
for i in all_stock_names:
    names.append(i)
```

### functions

```python
def omega_plus (P, gamma):
    result = (P**gamma)/(P**gamma+(1-P)**gamma)**(1/gamma)
    return result

def omega_minus (P, delta):
    result = (P**delta)/(P**delta+(1-P)**delta)**(1/delta)
    return result

def u_psilon_minus (x, labda, alpha):
    result = -labda * (-x)**alpha
    return result

def u_psilon_plus (x, alpha):
    result = x**alpha
    return result

def totalresults (nsigns, df):
    results = []
    for i in range(0, len(nsigns)):
        if i == 0:
            result = 0
            numbers = df['total'][0:nsigns[0]]
            for number in numbers:
                result += float(number)
            results.append(result)
        else:
            result = 0
            numbers = df['total'][nsigns[i-1]:nsigns[i-1]+nsigns[i]]
            for number in numbers:
                result += float(number)
            results.append(result)
    return results
```

### part 1 preperations

```python
tkval1 = []
tkval1b = []

for i in m:
        for l in range(-i,0):
            part1 = omega_minus((l+i+1)/60, 0.69) - omega_minus((l+i)/60, 0.69)
            tkval1.append(part1) 

for j in all_stock_names:
        for k in range(0, m[names.index(j)]):
            part1b = u_psilon_minus(df[j][k:k+1], 2.25, 0.88)            
            tkval1b.append(part1b)

part1 = pd.DataFrame(tkval1, columns = ['tkval1'])
part1.insert(1, "tkval1b" ,tkval1b, True)
part1['total'] = part1['tkval1'] * part1['tkval1b']            
```

### part 2 preperations

```python
tkval2 = []
tkval2b = []

for a in n:             
    for b in range(1,a+1):
        part2 = omega_plus((a-b+1)/60, 0.61) - omega_plus((a-b)/60, 0.61)
        tkval2.append(part2)
    
for j in all_stock_names:
        for k in range(60-n[names.index(j)],60):
            part2b = u_psilon_plus(df[j][k:k+1], 0.88)            
            tkval2b.append(part2b)   
            
part2 = pd.DataFrame(tkval2, columns = ['tkval2'])
part2.insert(1, "tkval2b" ,tkval2b, True)
part2['total'] = part2['tkval2'] * part2['tkval2b']
```

### Results

```python
results1 = totalresults (m, part1)
results2 = totalresults (n, part2)      
```

```python
allresults = pd.DataFrame(results1, columns = ['results1'])
allresults.insert(1, "results2", results2, True)
allresults['TK_values'] = allresults['results1'] + allresults['results2']
```

```python
allresults.insert(0,"stocks", names, True)
```
